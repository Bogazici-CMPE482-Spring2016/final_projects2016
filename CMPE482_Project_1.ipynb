#### 1) Read historical sales data and generates various plots like in Figure 1. The data will be a matrix in the workspace. You should automatically determine the number of items and the number of months in the data.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
```

```python
data_set_file = open ( 'data_set_1.txt' , 'r')
data_set_matrix = [ map(int,line.split(' ')) for line in data_set_file ]
for row in data_set_matrix:
    plt.plot(range(len(row)), row, '-o')
plt.xlabel('Month')
plt.ylabel('Number of Items Sold')
plt.title('Data Set 1')
plt.show()
```

#### 2)  Construct a basis regression model. When constructing your model, assume that there is a constant term, a linear trend and seasonal fluctuations (yearly, two-yearly and every 4 months).

```python
A = []
for column_index in range(len(data_set_matrix[0])):
    i = column_index
    A.append([1.0, 
              1.0*i,np.sin(0.5*np.pi*i),np.cos(0.5*np.pi*i),
              np.sin(np.pi*i/6),np.cos(np.pi*i/6),
              np.sin(np.pi*i/12),np.cos(np.pi*i/12)])
```

#### 3) For each least-squares method (1-Normal Equations, 2-QR factorization, 3-SVD based), your program will compute a prediction, using a basis regression model.

```python
x = np.transpose(data_set_matrix)
l = np.dot(np.transpose(A), A)
y_h = np.dot(np.transpose(A), x)
norm_eq = linalg.solve(l, y_h)

Q,R = linalg.qr(A, mode='economic')
qr_fact = linalg.lstsq(R,np.dot(np.transpose(Q),x))[0]

U,S,V = linalg.svd(A, full_matrices=False);
svd_dec = np.dot(np.dot(np.dot(np.transpose(V), np.diag(1 / S)),np.transpose(U)),x)

prediction_A = []
for column_index in range(len(data_set_matrix[0])+1, len(data_set_matrix[0])+7):
    i = column_index
    prediction_A.append([1.0, 
              1.0*i,np.sin(0.5*np.pi*i),np.cos(0.5*np.pi*i),
              np.sin(np.pi*i/6),np.cos(np.pi*i/6),
              np.sin(np.pi*i/12),np.cos(np.pi*i/12)])
import pandas
predictions_matrix = []
W_matrices = [norm_eq, qr_fact, svd_dec]
for i in range(3):
    predictions_matrix.append(np.dot(prediction_A, W_matrices[i]).reshape(-1))

total_relative_error = []
for i in range(3):
    arr = []
    for j in range(3):
        arr.append(linalg.norm(predictions_matrix[i][:]-predictions_matrix[j][:])/linalg.norm(predictions_matrix[i][:]))
    total_relative_error.append(arr)

from prettytable import PrettyTable
p = PrettyTable()
names = ["NormE","QR","SVD"]
p.add_row([""] + names)
for i, row in enumerate(total_relative_error):
    p.add_row([names[i]]+ row)
print "Total relative error between predictions generated by different computational methods"
print "NormE: Normal Equations"
print "QR : QR decomposition"
print "SVD : SVD Decomposition\n"
print p.get_string(header=False, border=False)
```
